```python
import numpy as np
```

```python
samples1=np.random.normal(loc=65,scale=5,size=(100,35))
samples2=np.random.normal(loc=30,scale=3,size=(100,35))
```

```python
samples_means1=np.mean(samples1,axis=1)
samples_means2=np.mean(samples2,axis=1)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(samples_means1,bins=11)
plt.xlabel('sample means of the sample withdrawn from n(65,25)')
plt.ylabel('frequency(sample mean=k)')
```

```python
plt.hist(samples_means2,bins=11)
plt.xlabel('sample means of the sample withdrawn from n(65,25)')
plt.ylabel('frequency(sample mean=k)')
```

```python
samples_means1
```

```python
samples_means2
```

```python
diff_bw_sample_means=samples_means1-samples_means2
```

```python
diff_bw_sample_means
```

```python
plt.hist(diff_bw_sample_means,bins=11)
plt.xlabel('diff b/w sample means of samples withdrawn from population 1 and population 2')
plt.ylabel('frequency(diff b/w samples_means1 and samples_means2 )')
```

```python
diff_mean=np.mean(diff_bw_sample_means)
```

```python
diff_mean
```

```python
np.var(diff_bw_sample_means)
```

```python
mu1_cap_best=np.mean(samples_means1)
```

```python
mu1_cap_best
```

```python
mu2_cap_best=np.mean(samples_means2)
```

```python
mu2_cap_best
```

```python
diff_best=mu1_cap_best-mu2_cap_best
```

```python
diff_best
```

```python
lcb=diff_best-(1.96*np.sqrt(np.var(diff_bw_sample_means)))
```

```python
ucb=diff_best+(1.96*np.sqrt(np.var(diff_bw_sample_means)))
```

```python
lcb
```

```python
ucb
```

```python
samples_from_diff_sd=np.random.normal(loc=diff_mean,scale=np.sqrt(np.var(diff_bw_sample_means)),size=(10,100))
```

```python
boolean_mask=(samples_from_diff_sd >lcb) & (samples_from_diff_sd < ucb)
```

```python
boolean_mask
```

```python
number_of_values_lying_in_ci=np.count_nonzero(boolean_mask,axis=1)
```

```python
number_of_values_lying_in_ci
```

```python
np.mean(number_of_values_lying_in_ci)
```

```python
one_sample1=np.random.normal(loc=65,scale=5,size=35)
```

```python
one_sample2=np.random.normal(loc=30,scale=3,size=35)
```

```python
mu1_best=np.mean(one_sample1)
```

```python
mu2_best=np.mean(one_sample2)
```

```python
var1_best=np.var(one_sample1)
```

```python
var2_best=np.mean(one_sample2)
```

```python
diff_best=mu1_best-mu2_best
```

```python
var_sd=(var1_best/35)+(var2_best/35)
```

```python
std_sd=np.sqrt(var_sd)
```

```python
lcb=diff_best-(1.96*std_sd)
```

```python
ucb=diff_best+(1.96*std_sd)
```

```python
lcb
```

```python
ucb
```

```python
samples=np.random.normal(loc=diff_best,scale=std_sd,size=(10,100))
```

```python
boolean_mask=(samples > lcb) & (samples < ucb)
```

```python
number_of_values=np.count_nonzero(boolean_mask,axis=1)
```

```python
number_of_values
```

```python
np.mean(number_of_values)
```
